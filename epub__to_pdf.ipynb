```python
# ==========================================
# 1️⃣ Install Required Tools
# ==========================================
!apt-get update -qq
!apt-get install -y -qq pandoc wkhtmltopdf

# ==========================================
# 2️⃣ Upload EPUB
# ==========================================
from google.colab import files
uploaded = files.upload()

# ==========================================
# 3️⃣ Safe Rename (Avoid Space & Special Char Issues)
# ==========================================
import os

original_name = list(uploaded.keys())[0]
safe_input = "input_book.epub"

os.rename(original_name, safe_input)

# ==========================================
# 4️⃣ Convert EPUB → PDF
# ==========================================
output_pdf = "converted_book.pdf"

# Execute pandoc command
!pandoc input_book.epub -o {output_pdf} --pdf-engine=wkhtmltopdf

# ==========================================
# 5️⃣ Download PDF
# ==========================================
# Check if the output PDF was actually created
if os.path.exists(output_pdf):
    files.download(output_pdf)
    print("✅ Conversion completed successfully!")
else:
    print(f"❌ Error: The file '{output_pdf}' was not created during conversion.")
    print("Please review the `pandoc` output above for any error messages or warnings.")
    print("The conversion from EPUB to PDF might have failed.")
```

```python
# ==========================================
# 1️⃣ Install Calibre
# ==========================================
!apt-get update -qq
!apt-get install -y -qq calibre

# ==========================================
# 2️⃣ Fix Chromium Sandbox Issue
# ==========================================
import os
os.environ["QTWEBENGINE_CHROMIUM_FLAGS"] = "--no-sandbox"

# ==========================================
# 3️⃣ Upload EPUB
# ==========================================
from google.colab import files
uploaded = files.upload()

original_name = list(uploaded.keys())[0]
safe_input = "book.epub"
os.rename(original_name, safe_input)

# ==========================================
# 4️⃣ Convert EPUB → PDF (A4 + Better Margins)
# ==========================================
!ebook-convert book.epub book.pdf \
--paper-size a4 \
--margin-top 25 \
--margin-bottom 25 \
--margin-left 25 \
--margin-right 25 \
--base-font-size 13

# ==========================================
# 5️⃣ Download PDF
# ==========================================
files.download("book.pdf")

print("✅ A4 PDF with extra margin created successfully!")
```
